```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**3
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
```

```python
axes = fig.add_axes([0,0,1,1])
```

```python
axes.plot(x,y, 'r')
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('plot')
fig
```

```python
fig, axes = plt.subplots(nrows = 1, ncols =2, figsize=(12,6), dpi=100)
```

```python
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
fig
```
